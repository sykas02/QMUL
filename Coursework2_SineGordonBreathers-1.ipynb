### 2.1 Background

The broad goal of the coursework is to use the techniques we have learned in the course to study the solutions to a particular non linear partial differential equation called [the Sine-Gordon equation](https://en.wikipedia.org/wiki/Sine-Gordon_equation). 

Once complete, the notebook should provide a self contained report on your numerical investigations and provide relevant commentary in markdown cells around it that would enable another student to understand, run and make basic modifications to the notebook.

The Sine Gordon equation is as follows:

$ \frac{\partial\varphi^2}{\partial t^2} -\frac{\partial\varphi^2}{\partial x^2}+\sin \varphi =0, \quad (1)$

and the breather solutions we will study have the form:

$ \varphi (x,t)=4\arctan \left({\frac {{\sqrt {1-\omega ^{2}}}\;\cos(\omega t)}{\omega \;\cosh({\sqrt {1-\omega ^{2}}}\;x)}}\right), \quad (2)$

with $\omega$ a positive constant with a value less than 1. In the coursework you should set $\omega=0.5$ unless otherwise specified.

Eq. (2) (the breather solution) can be used to construct initial data for the initial value problem for Eq. (1) (the PDE). If you correctly implement the time evolution of this data, you should find that it is stable - the evolution is described exactly by the breather solution, which oscillates in time but with the envelope of the function remaining stable. This provides a very good test that the code is working correctly, meaning that you can then confidently use it for different initial data for the same PDE, for which no exact solution exists. Implementing known solutions is a very good way to test a numerical code.

The following analytic results for the breather solution in Eq. (2) may be useful:

Let $A = \sqrt {1-\omega ^{2}}$, then

$ \frac{\partial\varphi (x,t)}{\partial x}= \frac{-8\omega A^2 \cos(\omega t)\sinh(Ax)}
{2 (A \cos(\omega t))^2 +  \omega^2 +  \omega^2 \cosh{2Ax}} \quad (3)$

$ \frac{\partial\varphi (x,t)}{\partial t}= \frac{-4 A \sin(\omega t)}
{\cosh{Ax} \left(1 + \left(\frac{A \cos{wt}}{\omega \cosh{Ax}}\right)^2 \right)} \quad (4)$

### Sine-Gordon Breather Solution Class
This class just implements methods to return values for given conditions and allows lagrange polynomial approximation of solutions

```python
class BreatherSolution:
    """
    BreatherSolution class offers a variety of methods to vary, calculate and approximate solutions to the Sine-Gordon equation (hyperbolic PDE)
    
    attributes: omega (parameter for breather solution), phi_output, lagrange_polynomial (most recently calculated polynomial for given x_values and t)
    
    methods: 
    calculate_output_phi(x, t) - calculates exact outputs for sine gordon breather solution in vectorized form
    calculate_x_derivative(x, t) - calculates exact outputs for x derivative of breather solution in vectorized form
    calculate_lagrange_polynomial(x_values, t) - given a set of nodes and t, it will calculate points for colocation with a lagrange polynomial (using sympy)

    
    """
    
    
    def __init__(self, omega=0.5):
        assert omega<1 and omega>0, 'Omega must be a positive constant less than 1'
        self.omega = omega
        
    def calculate_output_phi(self, x, t):
        """Computes vectorized calculations of sine-gordon breather solution
            x (float, int or ndarry), t (float or int) """
        assert isinstance(x, np.ndarray) or isinstance(x, float) or isinstance(x, int), 'x must be array, float or int'
        assert isinstance(t, float) or isinstance(t, int), 't must be float or int'
        
        output = 4 * np.arctan((np.sqrt(1 - (self.omega)**2) * np.cos(self.omega * t)) /
                               (self.omega * np.cosh(np.sqrt(1 - (self.omega)**2) * x)))
        self.phi_output = output
        
        return output
    
    def calculate_x_derivative(self, x, t):
        """return vectorized x derivative
            x (float, int or ndarry), t (float or int)"""
        assert isinstance(x, np.ndarray) or isinstance(x, float) or isinstance(x, int), 'x must be array, float or int'
        assert isinstance(t, float) or isinstance(t, int), 't must be float or int'
        A = np.sqrt(1-self.omega**2)
        value = ((-8*self.omega*(A**2)*np.cos(self.omega*t)*np.sinh(A*x)) / 
                 (2*(np.cos(self.omega*t)**2)+(self.omega**2)*(1+np.cosh(2*A*x))))
        return value
    
    def calculate_t_derivative(self, x, t):
        """return vectorized t derivative 
            x (float, int or ndarry), t (float or int)"""
        assert isinstance(x, np.ndarray) or isinstance(x, float) or isinstance(x, int), 'x must be array, float or int'
        assert isinstance(t, float) or isinstance(t, int), 't must be float or int'
        A = np.sqrt(1-self.omega**2)
        value = (-4*A*np.sin(self.omega*t))/(np.cosh(A*x)*((1+((A*np.cos(self.omega*t))/(self.omega*np.cosh(A*x)))**2)))
        return value
    
    def calculate_lagrange_polynomial(self, x_values, t):
        """Calculating lagrange polynomial from first principles using colocation of points and sympy"""
        assert isinstance(x_values, np.ndarray), 'x must be np.ndarray with n>0'
        assert isinstance(t, float) or isinstance(t, int), 't must be float or int'

        # calculating x, f(x) pairs for lagrange interpolation colocation
        y_values = self.calculate_output_phi(x_values, t=t)
        x = sympy.symbols('x') # using sympy to create an x variable for algebraic manipulation
        
        # initialise coefficients 
        P = 0
        # calculating lagrange polynomial from first principles
        for k, x_k in enumerate(x_values):
            # at the start of each k loop, L_k is initialised as 1 and each points' contribution is multiplied on
            L_k_k = 1
            for i, x_i in enumerate(x_values):
                if i != k:
                    L_k_k *= (x - x_i) / (x_k - x_i)
            P += L_k_k * y_values[k]
        
        self.lagrange_polynomial = sympy.simplify(P)
        return self.lagrange_polynomial
    
    def return_energy_equation(self, x, t):
        """Returns energy equation to be integrated as described in question 2, requires same x and t conditions as sinegordon solution"""
        return 1-np.cos(self.calculate_output_phi(x,t))

    
    def calculate_simpson_integral(self, func, a, b, N, t):
        assert isinstance(a, float) or isinstance(a,int),'a must be float or int'
        assert isinstance(b, float) or isinstance(b,int),'b must be float or int'
        assert isinstance(t, float) or isinstance(t,int),'t must be float or int'
        assert isinstance(N,int),'N must be an int'
        assert callable(func),'func must be a valid function with x, t input'
        
        x_values = np.linspace(a,b,N)
        y_values = func(x_values, t)
        integral_result = scipy.integrate.simpson(y_values, x_values)

        return integral_result
```

#### 2.2.3. Q3 Differentiation (22 marks for 2 parts (a)-(b)):

The code to generate the pseudospectral derivative matrices for the first and second derivatives of any function $f(x)$ for a grid containing $N$ points *at the Gauss Lobato points* in $x$ is given below for the standard interval $[-1,1]$, and verified on some example functions.

```python
# Code provided for the derivative matrices at Gauss Lobato nodes for the interval [-1,1]
# (For use in part 2.2.3)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Code provided for the derivative matrices at Gauss Lobato nodes for the interval [-1,1]

def get_pseudospectral_first_derivative_matrix(N) :
    
    indices = np.arange(N+1)
    u_i = np.cos(np.pi * indices / N)
    c_i = np.ones_like(indices)
    c_i[0] = 2.0
    c_i[N] = 2.0

    D_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == 0 and idx_j == 0) : 
                D_matrix[idx_i,idx_j] = (2.0 * N * N + 1.0)/6.0
                
            elif (idx_i == N and idx_j == N) :
                D_matrix[idx_i,idx_j] = -(2.0 * N * N + 1.0)/6.0
                
            elif (idx_i == idx_j) :
                D_matrix[idx_i,idx_j] = - u_i[idx_i] / 2.0 / (1.0 - u_i[idx_i] * u_i[idx_i])
                
            else :
                D_matrix[idx_i,idx_j] = (c_i[idx_i] / c_i[idx_j] * (-1)**(idx_i+idx_j) 
                                         / (u_i[idx_i] - u_i[idx_j]))
    return D_matrix

def get_pseudospectral_second_derivative_matrix(N) :
    
    indices = np.arange(N+1)
    u_i = np.cos(np.pi * indices / N)
    c_i = np.ones_like(indices)
    c_i[0] = 2.0
    c_i[N] = 2.0

    D2_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            if((idx_i == idx_j) and (idx_i == 0 or idx_i == N)) : 
                D2_matrix[idx_i,idx_j] = (N ** 4.0 - 1.0) / 15.0
            
            elif ((idx_i == idx_j) and (idx_i != 0) and (idx_i != N)) :
                D2_matrix[idx_i,idx_j] = (- ((N*N - 1.0) * (1.0 - u_i[idx_i] * u_i[idx_i]) + 3.0) 
                                             / (3.0 * (1.0 - u_i[idx_i] * u_i[idx_i]) ** 2.0))
                 
            elif (idx_i == 0 and idx_j > 0) :
                D2_matrix[idx_i,idx_j] = ((2.0 / 3.0) * ((-1) ** idx_j) / c_i[idx_j] 
                                          * ((2.0 * N * N + 1.0) * (1.0 - u_i[idx_j]) - 6.0)
                                          / (1.0 - u_i[idx_j]) / (1.0 - u_i[idx_j]))
                
            elif (idx_i == N and idx_j < N) :
                D2_matrix[idx_i,idx_j] = ((2.0 / 3.0) * ((-1) ** (N+idx_j)) / c_i[idx_j] 
                                          * ((2.0 * N * N + 1.0) * (1.0 + u_i[idx_j]) - 6.0)
                                          / (1.0 + u_i[idx_j]) / (1.0 + u_i[idx_j]))
                
            else :
                D2_matrix[idx_i,idx_j] = (((-1) ** (idx_i+idx_j)) / c_i[idx_j] 
                                            * (u_i[idx_i]*u_i[idx_i] + u_i[idx_i]*u_i[idx_j] - 2.0)
                                            / (1.0 - u_i[idx_i]*u_i[idx_i]) / (u_i[idx_i] - u_i[idx_j])
                                            / (u_i[idx_i] - u_i[idx_j]))
                
    return D2_matrix
```

**Q3b**. Write down the form of the second order accurate finite difference stencils for the 1st and 2nd derivatives in $x$, using a fixed grid spacing $h$. Explain (in words) how many points you need to include in each stencil, and how the coefficients can be found using Lagrange polynomials. Create the derivative matrix for a grid in $x$ containing $N$ points. Check them on the test functions and the breather solution in the same way as has been done for the pseudospectral derivative matrices above, and conclude what number of points is required for the derivative estimates at each point to match their true values to a good approximation by eye.

_HINT: Remember to use one sided derivatives at the ends of the matrix!_

Coefficients of the stencils can be found by fitting a lagrange polynomial to a set amount of points and then calculating the derivatives. For example to calculate second order accurate stencils we can first fit a lagrange polynomial to 3 points: $(x_1,f_1), (x_2,f_2) ,(x_3,f_3)$

The lagrange Polynomial is found using the formula implemented in Q1, which produces : $$P = \frac{f_1(x-x_2)(x-x_3)}{(x_1-x_2)(x_1-x_3)} + \frac{f_2(x-x_1)(x-x_3)}{(x_2-x_1)(x_2-x_3)} + \frac{f_3(x-x_1)(x-x_2)}{(x_3-x_1)(x_3-x_2)}$$
If we assume equal spacing i.e $h$ between $x_1, x_2, x_3$, The formula simplifies to: $$P = \frac{1}{2h^2}[f_1(x-x_2)(x-x_3)- 2 \cdot f_2(x-x_1)(x-x_3) + f_3(x-x_1)(x-x_2)] $$ 
Then differentiating this polynomial $$\frac{dP}{dx} = \frac{1}{2h^2}[f_1(2x-x_2-x_3)- 2 \cdot f_2(2x-x_1-x_3) + f_3(2x-x_1-x_2)] $$

Evaluating the Polynomial at x_2 for the interior derivative: $$\left.\frac{dP}{dx}\right|_{x=x_2} = \frac{1}{2h^2}[f_1(x_2-x_3)- 2 \cdot f_2(2x_2-x_1-x_3) + f_3(x_2-x_1)]$$

and given $x_{i+1}-x_i = h$ and $2x_2-x_1-x_3 = 0$ , therefore:

$$\left.\frac{dP}{dx}\right|_{x=x_2} = \frac{1}{2h}(f_3-f_1) + O(h^2)$$ 

For one sided stencils, you can subsitute x_1 and x_3 to get forward and backward stencils

$\left.\frac{dP}{dx}\right|_{x=x_1} = \frac{1}{2h}(-f_3+4f_2-3f_1) + O(h^2)$ , $\left.\frac{dP}{dx}\right|_{x=x_3} = \frac{1}{2h}(f_1-4f_2+3f_3) + O(h^2)$


Differentiating the Lagrange Polynomial again, we get the approximate for the second derivative central stencil.
$$\frac{d^2P}{dx^2} = \frac{1}{2h^2}[f_1(2)- 2 \cdot f_2(2) + f_3(2)] =  \frac{1}{h^2}[f_1 -2f_2 + f_3] $$

To get second order accurate one-sided stencils, you need to use 4 points in the lagrange polynomial, the same process would follow to differentiate twice and then substitute x_1 and x_4 for the stencils. Here are some reference stencils calculated https://www.colorado.edu/amath/sites/default/files/attached-files/wk10_finitedifferences.pdf

```python
def get_FD_first_derivative_matrix(N) :
    
    # splitting interval [-1,1] in to N points
    h = 2.0/N # h is spacing
    one_over_2h = 1.0 / (2*h)
    D_matrix = np.zeros([N+1,N+1])

    indices = np.arange(N+1)
    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == idx_j) :
                D_matrix[idx_i,idx_j] =  0

            elif (idx_i == (idx_j+1)) :
                D_matrix[idx_i,idx_j] =  -1 * one_over_2h
                
            elif (idx_i == (idx_j-1)) :
                D_matrix[idx_i,idx_j] =  1 * one_over_2h
                
            else :
                D_matrix[idx_i,idx_j] = 0.0
    
    # inputting boundary stencils
    D_matrix[0,0:3] = (np.array([-3,4,-1]) * one_over_2h)
    D_matrix[N,N-2:N+1] = (np.array([1,-4,3]) * one_over_2h)
    return D_matrix

def get_FD_second_derivative_matrix(N) :
    
    h = 2.0 / N
    one_over_h2 = 1.0 / (h*h) # 1/ (h squared)
    D2_matrix = np.zeros([N+1,N+1])
    indices = np.arange(N+1)

    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == idx_j) :
                D2_matrix[idx_i,idx_j] = -2 * one_over_h2
                
            elif (idx_i == (idx_j+1)) :
                D2_matrix[idx_i,idx_j] = 1 * one_over_h2
                
            elif (idx_i == (idx_j-1)) :
                D2_matrix[idx_i,idx_j] = 1 * one_over_h2
                
            else :
                D2_matrix[idx_i,idx_j] = 0.0
    # inputting boundary stencils
    D2_matrix[0,0:4]  = np.array([2,-5,4,-1]) * one_over_h2
    D2_matrix[N,N-3:N+1]=  np.array([-1, 4, -5, 2]) * one_over_h2
    return D2_matrix
```

### CFL

The CFL condition is a necessary but not sufficient condition on relative step size when solving PDE's numerically. In the case of our hyperbolic equation, the time step should satisfy $ \Delta t = \lambda \Delta x$ . 

We can find the $ \lambda $ by doing a Von Neumann stability analysis and introducing a harmonic perturbation solution and bounding the value so it does not diverge. Intuitively wave equations tend to have a $ \lambda \leq \frac{1}{c} $ and wave speed for the Sine-Gordon equation $ c=1 $ .  We can adjust boundary as needed to see convergence.


The solution is indeed stationary and not moving in the plane, it oscillates through time

```python
class SineGordonPDE :

    # Repurposed code from week 10 soln 
    
    """
    Represents the Sine Gordon Equation equation PDE and its solution
    
    attributes: a_x, a_y0 (initial conditions),
                N_grid (the size of the grid)
                a_method (integrating method used)
                _solution_t , _solution_y stores values from integrating
    
    """
    
    # constructor function
    def __init__(self, a_x, a_y0, a_max_time, a_CFL=1.0, a_method="finite_difference"):
        self.CFL = a_CFL
        self.max_time = a_max_time
        self.x = a_x
        self.y0 = a_y0
        self.N_grid = np.size(a_x)            
        self._solution_t = []
        self._solution_y = []
        self.method = a_method
        
        self.initial_value = self.y0
        
        if self.method == "finite_difference" :
            self.D2_matrix = get_FD_second_derivative_matrix(self.N_grid-1)
        elif self.method == "pseudospectral" :
            self.D2_matrix = get_pseudospectral_second_derivative_matrix(self.N_grid-1)
        else :
            assert False, "Requested derivative method not implemented"
            
        # Rescale for the derivative matrix for the interval we are looking at
        b_minus_a = self.x[self.N_grid-1] - self.x[0]
        self.D2_matrix *= 4.0 / b_minus_a / b_minus_a

    def calculate_dydt(self, t, current_state) :
        
        # Just for readability
        [u,v] = np.array_split(current_state, 2)
        dydt = np.zeros_like(current_state)
        dudt, dvdt = np.array_split(dydt, 2)
        
        # Now work out the time derivatives 
        dudt[:] =  v
        dvdt[:] =  np.dot(self.D2_matrix, u) - np.sin(u) # sine gordon
           
        # Zero the derivatives at the end for stability 
        dydt[0] = 0.0
        dydt[1] = 0.0
        dydt[self.N_grid-1] = 0.0
        dydt[self.N_grid-2] = 0.0
        dydt[self.N_grid] = 0.0
        dydt[self.N_grid+1] = 0.0
        dydt[self.N_grid*2-1] = 0.0
        dydt[self.N_grid*2-2] = 0.0
        
        return dydt      
 
    def integrate_using_solve_ivp(self) :
        
        t_solution = np.linspace(0.0, self.max_time, 100)
        solution = scipy.integrate.solve_ivp(self.calculate_dydt, [0, self.max_time],
                        self.initial_value, t_eval=t_solution)
        
        self._solution_t = solution.t
        self._solution_y = solution.y
        
        self._solution_solved = True
        
        return solution.t, solution.y
    
    def integrate_using_midpoint(self) :
        
        assert self.method == "finite_difference", 'need to use FDs'
        dx = self.x[1] - self.x[0] # assume constant using FDs
        
        # Using half the CFL condition in practise just to be sure it is stable
        delta_t = self.CFL * dx
        num_timesteps = int(self.max_time / delta_t) + 1
        t_solution = np.linspace(0.0, self.max_time, num_timesteps)
        y_solution = np.zeros([2*self.N_grid, num_timesteps])
        
        for itime, time in enumerate(t_solution) :
            if itime == 0 :
                y_solution[:, itime] = self.initial_value
                old_time = time
            else :
                dydt_at_t = self.calculate_dydt(old_time, y_solution[:, itime-1])
                
                # Take half step
                y_t_plus_half = y_solution[:, itime-1] + 0.5 * dydt_at_t * delta_t
                t_plus_half = old_time + 0.5 * delta_t
                dydt_at_t_plus_half = self.calculate_dydt(t_plus_half, y_t_plus_half)
                
                # Take full step
                y_solution[:, itime] = (y_solution[:, itime-1] + dydt_at_t_plus_half * delta_t) 
                old_time = time 

        self._solution_t = t_solution
        self._solution_y = y_solution
        
        return t_solution, y_solution
    
    def integrate_using_implicit_method(self, a_num_steps=1000) :
        
        t_solution = np.linspace(0.0, self.max_time, a_num_steps)
        y_solution = np.zeros([2*self.N_grid, a_num_steps])
        delta_t = self.max_time / (a_num_steps - 1)
        
        for itime, time in enumerate(t_solution) :
            if itime == 0 :
                y_solution[:,itime] = self.initial_value
            else :
                y_of_t_old   = y_solution[:,itime-1]
                
                y_of_t_guess = np.copy(y_of_t_old)
                
                # Implement the iterative scheme described in the lecture
                error = 10.0
                error_threshold = 1e-1
                while (error > error_threshold) :
                    
                    # Get the value of dydt at the guessed value
                    dydt_at_t_next = self.calculate_dydt(time, y_of_t_guess)
                    
                    # Get the new guess using the trapezoidal update rule
                    y_of_t_new = y_of_t_old + (delta_t/2) * (dydt_at_t_next + self.calculate_dydt(t_solution[itime-1],y_of_t_old))
                    
                    # If the new guess is the same as the old one we have converged
                    error = np.linalg.norm(y_of_t_guess - y_of_t_new)
                    y_of_t_guess = y_of_t_new
                    
                    # Check it is not getting worse!
                    assert error < 20.0, 'Non linear iterations not converging!'
                    
                # Now (assuming it has converged) assign the 
                # value we found to the solution and continue the time iteration
                # Enforce the boundary conditions
                y_of_t_guess[0] = 0.0
                y_of_t_guess[self.N_grid-1] = 0.0
                y_solution[:,itime] = y_of_t_guess
                
        self._solution_t = t_solution
        self._solution_y = y_solution
    
        self.check_solution_converged(a_num_steps, delta_t)
        return t_solution, y_solution


    def check_solution_converged(self, a_num_steps, a_delta_t) :
        
        # Check whether the solution is converged
        last_but_one_solution = self._solution_y[:,a_num_steps-2]
        last_solution = self._solution_y[:,a_num_steps-1]
        
        change_at_last_step = last_solution - last_but_one_solution
        final_difference = np.linalg.norm(change_at_last_step, 2)/a_delta_t
        is_converged = (final_difference < 4.0)
        print("The error in the final solution is", final_difference) 
        assert is_converged, 'solution not converged, try increasing max_time'
        
        self._solution_solved = True
        
```

```python
N = 100
a = -15
b = 15
x = np.linspace(a,b,N)
u0 = breather.calculate_output_phi(x, 0)  
v0 = breather.calculate_t_derivative(x, 0)
y0 = np.concatenate([u0,v0])
max_time = 200.0

wave = SineGordonPDE(x, y0, max_time, a_CFL=1.0)


t_solution, y_solution = wave.integrate_using_implicit_method()

u_solution = y_solution[0:N,:]

for i, t_i in enumerate(t_solution) :
    if (i > 0) and (i % 5 == 0) and (t_i > 0.0):
        labelt = "t="+str(round(t_i,2))
        f_t = u_solution[:,i]
        plt.plot(x, f_t, label=labelt, alpha=(1-t_i/max_time))

plt.title('Implicit Trapezoidal with FD derivatives solution')
plt.xlabel('x')
plt.ylabel('value over time of u')
plt.grid()
# plt.legend(loc='upper right')
plt.show()
```

## advantages v disadvantages

adv - unconditionally stable and good at handling stiff problems where explicit methods may require very small time steps to converge


disadv - more complex computations required for implicit methods and requires iterative methods to reduce errors which is costly

dependent on time step size for convergence and may take time with trial and error to work out correctly
